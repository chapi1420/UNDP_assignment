```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
normal_data = np.random.normal(loc=50, scale=10, size=1000)
uniform_data = np.random.uniform(low=0, high=100, size=1000)
poisson_data = np.random.poisson(lam=5, size=1000)
exponential_data = np.random.exponential(scale=10, size=1000)
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Plot each distribution
sns.histplot(normal_data, bins=30, kde=True, ax=axes[0, 0]).set_title("Normal Distribution")
sns.histplot(uniform_data, bins=30, kde=True, ax=axes[0, 1]).set_title("Uniform Distribution")
sns.histplot(poisson_data, bins=30, kde=True, ax=axes[0, 2]).set_title("Poisson Distribution")
sns.histplot(exponential_data, bins=30, kde=True, ax=axes[1, 0]).set_title("Exponential Distribution")
sns.histplot(binomial_data, bins=30, kde=True, ax=axes[1, 1]).set_title("Binomial Distribution")

# Remove empty subplot
fig.delaxes(axes[1, 2])

plt.tight_layout()
plt.show()
```
